## Lets filter out Recent data

```python
recent_data = netflix[netflix['release_year'] >= 2014]
recent_data.head()
```

```python
print('Movies : {}'.format(len(recent_data[recent_data['type']=='Movie'])))
print('TV Shows : {}'.format(len(recent_data[recent_data['type']=='TV Show'])))
```

```python
recent_data1 = netflix[netflix['release_year'] >= 2017]
recent_data1.head()
```

```python
print('Movies : {}'.format(len(recent_data1[recent_data1['type']=='Movie'])))
print('TV Shows : {}'.format(len(recent_data1[recent_data1['type']=='TV Show'])))
```

```python
diff = netflix.groupby('release_year')['type'].agg(lambda x: x[x == 'Movie'].count() - x[x == 'TV Show'].count())
diff.name = 'diff'
diff
```

```python
netflix[netflix['release_year']==1942]
```

```python
diff.plot(figsize=(20,8), xticks=diff.index, ylabel='count', rot=90)
```

```python
normalized_diff = (diff - diff.mean())/diff.std()
normalized_diff.plot(figsize=(20,8), xticks=normalized_diff.index, ylabel='count', rot=90)
```

```python
normalized_diff[normalized_diff.index>2010]
```

```python
netflix[(netflix.type == 'TV Show') & (netflix.release_year >2010)].groupby('release_year')['show_id'].count()
```

```python
netflix[(netflix.type == 'Movie') & (netflix.release_year >2010)].groupby('release_year')['show_id'].count()
```

- So we can conclude that there is sharpe change from movies to TV shows. Last 5 years there is increase in tv shows percentage
- while decrease in Movies.
